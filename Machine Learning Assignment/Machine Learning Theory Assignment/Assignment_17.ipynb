**Ans:** The equation `y=mx+c` represents a straight line graphically, where m is its slope/gradient and c its intercept. In this tutorial, you will learn how to plot y=mx+b in Python with Matplotlib.

### Q2. In a graph, explain the terms rise, run, and slope ?

**Ans:** 
```python
'Rise:' The "rise" refers to the vertical change between two points on a graph. It represents how much the dependent variable (y-axis) changes between two corresponding values of the independent variable (x-axis).

'Run:' The "run" refers to the horizontal change between two points on a graph. It represents how much the independent variable (x-axis) changes between two corresponding values of the dependent variable (y-axis).

'Slope:' The slope of a line on a graph is the ratio of the rise to the run. Mathematically, it's defined as the change in the y-coordinate divided by the change in the x-coordinate between two points on the line.
```

### Q3. Use a graph to demonstrate slope, linear positive slope, and linear negative slope, as well as the different conditions that contribute to slope ?

**Ans:** The slope of a line on a graph is the ratio of the rise to the run. Mathematically, it's defined as the change in the y-coordinate divided by the change in the x-coordinate between two points on the line.

`Slope=Rise/Run=ChangeinY/changeinX`

The slope of a line is usually represented by the letter m. (x1, y1) represents the first point whereas (x2, y2) represents the second point.  

`M=Y2−Y1/X2−X1`

It is important to keep the x-and y-coordinates in the same order in both the numerator and the denominator otherwise you will get the wrong slope.


### Q4. Use a graph to demonstrate curve linear negative slope and curve linear positive slope ?

**Ans:** 
**Curve Linear Negative Slope:** In this case, the curve will slope downward from left to right.<br>
**Curve Linear Positive Slope:** In this case, the curve will slope upward from left to right.<br>
 

y=ax^2+bx+c to create these curves,<br> 
**where:-**<br>

**a**:- determines the direction and steepness of the curve.<br>
**b**:- adjusts the position of the curve along the x-axis.<br>
**c**:- is the y-intercept, where the curve intersects the y-axis.



````python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 100)

# Parameters for curve linear negative slope
a_neg = -0.5
b_neg = 0
c_neg = 3

# Parameters for curve linear positive slope
a_pos = 0.5
b_pos = 0
c_pos = -3

# Calculate y values for both curves
y_neg = a_neg * x**2 + b_neg * x + c_neg
y_pos = a_pos * x**2 + b_pos * x + c_pos

# Plot curve linear negative slope
plt.plot(x, y_neg, label='Curve Linear Negative Slope', color='red')

# Plot curve linear positive slope
plt.plot(x, y_pos, label='Curve Linear Positive Slope', color='blue')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve Linear Negative and Positive Slope')
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 100)

# Parameters for the curve
a = 0.5
b = 0
c = -3

# Calculate y values
y = a * x**2 + b * x + c

# Calculate the derivative dy/dx
dy_dx = 2 * a * x + b

# Calculate the second derivative d^2y/dx^2
d2y_dx2 = 2 * a

# Find the critical points where dy/dx = 0
critical_points = -b / (2 * a)

# Calculate the y values at critical points
y_critical = a * critical_points**2 + b * critical_points + c

# Plot the curve
plt.plot(x, y, label='Curve')

# Plot critical points
plt.plot(critical_points, y_critical, 'ro', label='Critical Point')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve with Maximum and Minimum Points')
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
````

### Q10. Describe the regression analysis assumptions and the BLUE principle ?

**Ans:** There are four assumptions associated with a linear regression model:
- **Linearity:** The relationship between X and the mean of Y is linear.
- **Homoscedasticity:** The variance of residual is the same for any value of X. 
- **Independence:** Observations are independent of each other.
- **BLUE:** is an acronym for the following: Best Linear Unbiased Estimator. In this context, the definition of “best” refers to the minimum variance or the narrowest sampling distribution.
